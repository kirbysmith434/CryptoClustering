```python

# Import required libraries and dependencies
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id"
)

# Display sample data
df_market_data.head(10)

# Normalize the data using StandardScaler
scaler = StandardScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df_market_data), columns=df_market_data.columns, index=df_market_data.index)

# Create a DataFrame with the scaled data
df_scaled_data = df_scaled.copy()

# Check if 'crypto_name' is present in the original data
if 'crypto_name' in df_market_data.columns:
    # Copy the 'crypto_name' column from the original data
    df_scaled_data['crypto_name'] = df_market_data['crypto_name']
else:
    # If 'crypto_name' is not present, you can set it to be the index or use some other identifier
    df_scaled_data['crypto_name'] = df_market_data.index

# Reset the index to avoid the 'KeyError' during the next step
df_scaled_data.reset_index(inplace=True)

# Set the 'coin_id' column as index
df_scaled_data.set_index('coin_id', inplace=True)

# Display sample data
df_scaled_data.head(10)
```

---

### Find the Best Value for k Using the Original Data.

```python
# Create a list with the number of k-values from 1 to 11
k_values = list(range(1, 12))

# Display the list
print(k_values)
```

```python
# Import required libraries and dependencies
from sklearn.cluster import KMeans

# Create an empty list to store the inertia values
inertia_values = []

# Create a for loop to compute the inertia with each possible value of k
for k in range(1, 12):
    # 1. Create a KMeans model using the loop counter for the n_clusters
    kmeans_model = KMeans(n_clusters=k, random_state=42, n_init=10)  # Explicitly set n_init

    # 2. Fit the model to the data using `df_market_data_scaled`
    kmeans_model.fit(df_scaled)

    # 3. Append the model.inertia_ to the inertia list
    inertia_values.append(kmeans_model.inertia_)

# Display the inertia values
print(inertia_values)

```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data = {'k_values': list(range(1, 12)),
              'inertia_values': inertia_values}

# Display the dictionary
print(elbow_data)

# Create a DataFrame with the data to plot the Elbow curve
df_elbow = pd.DataFrame(elbow_data)

# Display the DataFrame
df_elbow
```

```python
# Import required libraries and dependencies
import matplotlib.pyplot as plt

# Plot a line chart with inertia values
plt.figure(figsize=(10, 6))
plt.plot(df_elbow['k_values'], df_elbow['inertia_values'], marker='o', linestyle='-', color='b')
plt.title('Elbow Curve for Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.grid(True)
plt.show()
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** 4

### Cluster Cryptocurrencies with K-means Using the Original Data

```python
# Assuming the optimal value for k is 4
optimal_k = 4

# Initialize the K-Means model with the best value for k
best_kmeans_model = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)

# Fit the model to the scaled data
best_kmeans_model.fit(df_scaled)

# Get the cluster labels
cluster_labels = best_kmeans_model.labels_

# Display the cluster labels
print(cluster_labels)

```

```python
# Initialize the K-Means model without specifying the number of clusters
kmeans_model = KMeans(random_state=42, n_init=10)

# Fit and predict clusters on the scaled data
cluster_labels = kmeans_model.fit_predict(df_scaled)

# Display the cluster labels
print(cluster_labels)

```

```python
# Predict the clusters to group the cryptocurrencies using the scaled data
cluster_labels = best_kmeans_model.predict(df_scaled)

# Print the resulting array of cluster values
print(cluster_labels)

```

### Optimize Clusters with Principal Component Analysis.

```python
# Import required libraries and dependencies
from sklearn.decomposition import PCA

# Create a PCA model instance with n_components=3
pca_model = PCA(n_components=3)
```

```python
# Apply PCA to the scaled data
pca_result = pca_model.fit_transform(df_scaled)

# Create a DataFrame with the PCA result
df_pca = pd.DataFrame(data=pca_result, columns=['PC1', 'PC2', 'PC3'], index=df_scaled.index)

# Display the PCA result DataFrame
df_pca.head()
```

```python
# Import required libraries and dependencies
from sklearn.decomposition import PCA

# Create a PCA model instance with n_components=3
pca_model = PCA(n_components=3)

# Fit the PCA model to the scaled data
pca_result = pca_model.fit_transform(df_scaled)

# Retrieve the explained variance
explained_variance = pca_model.explained_variance_ratio_

# Display the explained variance for each principal component
for i, variance in enumerate(explained_variance, start=1):
    print(f'Explained Variance PC{i}: {variance:.4f}')
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** 0.895

```python
# Creating a DataFrame with the PCA data
df_pca_data = pd.DataFrame(data=pca_result, columns=['PC1', 'PC2', 'PC3'], index=df_scaled.index)

# Copy the crypto names from the original data
df_pca_data['crypto_name'] = df_scaled_data['crypto_name']

# Set the index
df_pca_data.index.name = 'coin_id'

# Display sample data
df_pca_data.head()

```

### Find the Best Value for k Using the PCA Data

```python
# Create a list with the number of k-values from 1 to 11
k_values = list(range(1, 12))

# Display the list
print(k_values)

```

```python
# Create an empty list to store the inertia values
inertia_values = []

# Drop the categorical column 'crypto_name' before applying KMeans
df_pca_data_numeric = df_pca_data.drop('crypto_name', axis=1)

# Create a for loop to compute the inertia with each possible value of k
for k in k_values:
    # 1. Create a KMeans model using the loop counter for the n_clusters
    kmeans_model = KMeans(n_clusters=k, random_state=42, n_init=10)

    # 2. Fit the model to the numeric data using `df_pca_data_numeric`
    kmeans_model.fit(df_pca_data_numeric)

    # 3. Append the model.inertia_ to the inertia list
    inertia_values.append(kmeans_model.inertia_)

# Display the inertia values
print(inertia_values)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data = {'k_values': k_values, 'inertia_values': inertia_values}

# Create a DataFrame with the data to plot the Elbow curve
df_elbow = pd.DataFrame(elbow_data)

# Display the DataFrame
df_elbow
```

```python
# Import required libraries and dependencies
import hvplot.pandas

# Plot a line chart with all the inertia values
df_elbow.hvplot.line(
    x='k_values',
    y='inertia_values',
    title='Elbow Curve',
    xlabel='Number of Clusters (k)',
    ylabel='Inertia',
    width=800,
    height=500,
    line_width=2
)
```

  * **Answer:** 4


* **Question:** Does it differ from the best k value found using the original data?

  * **Answer:** No

### Cluster Cryptocurrencies with K-means Using the PCA Data

```python
# Initialize the K-Means model using the best value for k
best_k = 4
kmeans_model = KMeans(n_clusters=best_k, random_state=42, n_init=10)
```

```python
# Fit the K-Means model using the PCA data
kmeans_model.fit(df_pca_data_numeric)
```

```python
# Predict the clusters to group the cryptocurrencies using the PCA data
cluster_labels = kmeans_model.predict(df_pca_data_numeric)

# Print the resulting array of cluster values
print(cluster_labels)

```

```python
# Create a copy of the DataFrame with the PCA data
df_pca_with_clusters = df_pca_data_numeric.copy()

# Add a new column to the DataFrame with the predicted clusters
df_pca_with_clusters['cluster_label'] = cluster_labels

# Display sample data
df_pca_with_clusters.head()

```

```python
# Create a scatter plot using hvPlot
scatter_plot = df_pca_with_clusters.hvplot.scatter(
    x='PC1',
    y='PC2',
    by='cluster_label',
    hover_cols=['crypto_name'],
    title='Scatter Plot of PCA Data with K-Means Clusters',
    xlabel='PC1',
    ylabel='PC2',
    width=800,
    height=500
)

# Show the scatter plot
scatter_plot
```
